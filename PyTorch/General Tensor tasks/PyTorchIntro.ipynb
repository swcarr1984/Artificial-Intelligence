```python
# Import torch and check if successful by verifying version
import torch
torch.__version__
```

```python
# Scalar
scalar = torch.tensor(7)
print(scalar,scalar.ndim,scalar.item())
```

```python
# Vector
vector = torch.tensor([7,7])
print(vector,vector.ndim,vector.shape)
```

```python
# Matrix - normally capitalised
MATRIX = torch.tensor([[7,8], 
                       [9,10]])
print(MATRIX.ndim,",",MATRIX.shape)
```

```python
# Tensor - normally capitalised
TENSOR = torch.tensor([[[1,2,3]
                        [3,6,9]
                        [2,4,5]]])
# Check dimensions and shape of tensor
print(TENSOR.ndim,",",TENSOR.shape
```

```python
# Creating random tensors of certain sizing
rand_tensor = torch.rand(size=(3,4))
rand_tensor, rand_tensor.dtype
```

```python
# Zero tensors
zeros =  torch.zeros(size=(3,4))
zeros, zeros.dtype
```

```python
# Tensor of ones
ones = torch.ones(size=(3,4))
ones, ones.dtype
```

```python
# Creating range of numbers 0-40 in 10's
first_40 = torch.arange(start=0, end=50, step=10)
first_40
```

```python
# Copy shape and load with new values - use previous tensor
zeroed_first_40 = torch.zeros_like(input=first_40)
zeroed_first_40
```

```python
# Check where tensor is stored
new_tensor = torch.rand(2,3)
print(new_tensor)
print(f"This tensor is stored on: {new_tensor.device}, with datatype: {new_tensor.dtype}")
```
