```python
import torch
```

```python
# Create a tensor of values and add a number to it
tensor = torch.tensor([1,2,3])
add_to_tensor = tensor+10
multiply_tensor = tensor*10 # Can also use -> torch.multiply(tensor,10)
# Reassign original tensor
tensor = tensor+10
# Print all three results
print(add_to_tensor, multiply_tensor,tensor)
```

```python
# Element-wise and Matrix Multiplication
my_tensor = torch.tensor([1,2,3])
# Element-wise matrix multiplication
tensor_result1 = my_tensor*my_tensor
# Matrix multiplication
tensor_result2 = torch.matmul(my_tensor,my_tensor)
tensor_result3 = torch.matmul
print(tensor_result1, tensor_result2)
```

```python
# Perform Transpose on Matrix to allow multiplication of same sizes
new_tensor1 = torch.tensor([[1,2],
                            [3,4],
                            [5,6]], dtype=torch.float32)

new_tensor2 = torch.tensor([[7,10],
                            [8,11],
                            [9,12]], dtype=torch.float32)
print(new_tensor2.T)

result_tensor = torch.matmul(new_tensor1, new_tensor2.T)
print(result_tensor)
```
